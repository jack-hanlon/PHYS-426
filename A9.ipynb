```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
#Define Variables
U = 1 #Let U = 1 for simplification
H = 1 #Let H = 1 for simplification
v = 0.1 #Thus viscosity will be 0.1
t_norm = [0.01,0.05,0.2,0.5,1.,10.]
t = t_norm

z = np.linspace(0,1,100)
w0 = []
w1 = []
w2 = []
w3 = []
w4 = []
w5 = []
t0 = t[0]
t1 = t[1]
t2 = t[2]
t3 = t[3]
t4 = t[4]
t5 = t[5]
for i in range(0,len(z)):
    w0.append((2*U/np.pi)*-1*(np.exp(-1*np.pi*np.pi*v*t0/H**2))*np.sin(np.pi*z[i]/H))
    w1.append((2*U/np.pi)*-1*(np.exp(-1*np.pi*np.pi*v*t1/H**2))*np.sin(np.pi*z[i]/H))
    w2.append((2*U/np.pi)*-1*(np.exp(-1*np.pi*np.pi*v*t2/H**2))*np.sin(np.pi*z[i]/H))
    w3.append((2*U/np.pi)*-1*(np.exp(-1*np.pi*np.pi*v*t3/H**2))*np.sin(np.pi*z[i]/H))
    w4.append((2*U/np.pi)*-1*(np.exp(-1*np.pi*np.pi*v*t4/H**2))*np.sin(np.pi*z[i]/H))
    w5.append((2*U/np.pi)*-1*(np.exp(-1*np.pi*np.pi*v*t5/H**2))*np.sin(np.pi*z[i]/H))
```

```python
plt.plot(z,w0,label="t = 0.01 ")
plt.plot(z,w1,label="t = 0.05")
plt.plot(z,w2,label="t = 0.2")
plt.plot(z,w3,label="t = 0.5")
plt.plot(z,w4,label="t = 1.0")
plt.plot(z,w5,label="t = 10.0")
plt.title("w(z,t) solution")
plt.xlabel("z")
plt.ylabel("w(z)")
plt.legend()
```

```python
u0 = []
u1 = []
u2 = []
u3 = []
u4 = []
u5 = []
t0 = t[0]
t1 = t[1]
t2 = t[2]
t3 = t[3]
t4 = t[4]
t5 = t[5]
for i in range(0,len(z)):
    u0.append((z[i]/H)+ (2*U/np.pi)*-1*(np.exp(-1*np.pi*np.pi*v*t0/H**2))*np.sin(np.pi*z[i]/H))
    u1.append((z[i]/H)+ (2*U/np.pi)*-1*(np.exp(-1*np.pi*np.pi*v*t1/H**2))*np.sin(np.pi*z[i]/H))
    u2.append((z[i]/H)+ (2*U/np.pi)*-1*(np.exp(-1*np.pi*np.pi*v*t2/H**2))*np.sin(np.pi*z[i]/H))
    u3.append((z[i]/H)+ (2*U/np.pi)*-1*(np.exp(-1*np.pi*np.pi*v*t3/H**2))*np.sin(np.pi*z[i]/H))
    u4.append((z[i]/H)+ (2*U/np.pi)*-1*(np.exp(-1*np.pi*np.pi*v*t4/H**2))*np.sin(np.pi*z[i]/H))
    u5.append((z[i]/H)+ (2*U/np.pi)*-1*(np.exp(-1*np.pi*np.pi*v*t5/H**2))*np.sin(np.pi*z[i]/H))
```

```python
plt.plot(z,u0,label="t = 0.01 ")
plt.plot(z,u1,label="t = 0.05")
plt.plot(z,u2,label="t = 0.2")
plt.plot(z,u3,label="t = 0.5")
plt.plot(z,u4,label="t = 1.0")
plt.plot(z,u5,label="t = 10.0")
plt.title("u(z,t) solution")
plt.xlabel("z")
plt.ylabel("u(z)")
plt.legend()
```
